```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
import sys
%matplotlib inline
```

```python
# Read the original image
image = cv2.imread('05.png')
plt.imshow(image)
```

```python
import cv2
# Parameters: src, dst, h_luminance (10), hColor (10), templateWindowSize (7), searchWindowSize (21)
dst = cv2.fastNlMeansDenoisingColored(image, None, 10, 10, 7, 21)
plt.imshow(dst)
```

```python
# Convert to grayscale and Gaussian blur
gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
gray_image = cv2.fastNlMeansDenoisingColored(gray_image, None, 15, 21)
gray_image = cv2.GaussianBlur(gray_image, (13, 13), 13)
plt.imshow(gray_image, cmap="gray")
```

```python
# Adaptive thresholding
thresh_image = cv2.adaptiveThreshold(gray_image, 255, cv2.ADAPTIVE_THRESH_GAUSSIAN_C, cv2.THRESH_BINARY, 17, 2)
plt.imshow(thresh_image, cmap="gray")
```
